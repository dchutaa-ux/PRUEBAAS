```python
def operaciones(a, b):
    return a+b, a-b, a*b, a/b

resultado = operaciones(10, 2)
print(resultado)        # (12, 8, 20, 5.0)

suma, resta, prod, div = operaciones(10, 2)
print("Suma:", suma)
print("División:", div)
```

```python
columnas = "ABCDEFGH"
for fila in range(1, 9):
    for col in range(8):
        print(f"{columnas[col]}{fila}", end=" ")
    print()
```

```python
texto = "python"
desplazamiento = 3
abecedario = "abcdefghijklmnopqrstuvwxyz"

cifrado = ""
for letra in texto:
    idx = (abecedario.index(letra) + desplazamiento) % 26
    cifrado += abecedario[idx]

print("Texto cifrado:", cifrado)
```
